generate synthetic iot dataset

```
np.random.seed(42)

samples = 1000

temperature = np.random.normal(35, 3, samples)

pressure = np.random.normal(110, 5, samples)

humidity = np.random.normal(60, 8, samples)

vibration = np.random.normal(4, 0.7, samples)

current = np.random.normal(20, 2, samples)

df = pd.DataFrame({
    "Temperature": temperature,
    "Pressure": pressure,
    "Humidity": humidity,
    "Vibration": vibration,
    "Current": current
})

print(df.head())
```

```
scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_data,
                         columns=df.columns)
```

convert into time series sequence

```
import numpy as np
import pandas as pd

def create_sequences(data_array, seq_len):
    xs = []
    for i in range(len(data_array) - seq_len + 1):
        xs.append(data_array[i:(i + seq_len)])
    return np.array(xs)

# Load the saved scaled data
scaled_df = pd.read_csv('iot_sensor.csv')

# Convert DataFrame to numpy array for sequence creation
data_array = scaled_df.values

# Create time series sequences
data = create_sequences(data_array, seq_len)

print(f"Shape of the generated time series data: {data.shape}")
```

define timegan parameter

```
gan_args = {
    "batch_size":128,
    "lr":5e-4,
    "noise_dim":32,
    "layers_dim":128
}
```

create timegan model

```
import torch
import torch.nn as nn
import torch.optim as optim

# Assuming `data` and `seq_len` are already defined from previous steps
# For demonstration, let's assume 'data' is a torch tensor
data_tensor = torch.from_numpy(data).float() # Convert numpy array to torch tensor

# Define the dimensions based on your data
feature_dim = data_tensor.shape[-1] # Number of features per time step
hidden_dim = gan_args['layers_dim']
noise_dim = gan_args['noise_dim']

# --- Step 1: Define the TimeGAN model components (Placeholders) ---

# In practice, these would be custom nn.Module classes implementing
# the GRU-based networks described in the TimeGAN paper.

class Embedder(nn.Module):
    def __init__(self, input_dim, hidden_dim):
        super(Embedder, self).__init__()
        # Embedder GRU maps feature_dim to hidden_dim for each timestep
        self.gru = nn.GRU(input_dim, hidden_dim, batch_first=True)

    def forward(self, x):
        # x: (batch_size, seq_len, feature_dim)
        output, _ = self.gru(x) # output: (batch_size, seq_len, hidden_dim)
        return output # Return the full sequence of embedded hidden states

class Recovery(nn.Module):
    def __init__(self, hidden_dim, output_dim):
        super(Recovery, self).__init__()
        # Recovery GRU maps hidden_dim (embedded sequence) back to hidden_dim
        self.gru = nn.GRU(hidden_dim, hidden_dim, batch_first=True)
        # Final FC layer maps hidden_dim to original feature_dim
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, embedded_sequence): # embedded_sequence: (batch_size, seq_len, hidden_dim)
        gru_output, _ = self.gru(embedded_sequence) # gru_output: (batch_size, seq_len, hidden_dim)
        recovered_sequence = self.fc(gru_output) # (batch_size, seq_len, feature_dim)
        return recovered_sequence

class Generator(nn.Module):
    def __init__(self, noise_dim, hidden_dim):
        super(Generator, self).__init__()
        # Generator GRU takes noise_dim and generates a hidden_dim sequence
        self.gru = nn.GRU(noise_dim, hidden_dim, batch_first=True)
        # Final FC layer maps hidden_dim to hidden_dim (the embedded space)
        self.fc = nn.Linear(hidden_dim, hidden_dim)

    def forward(self, z): # z: (batch_size, seq_len, noise_dim)
        output, _ = self.gru(z) # output: (batch_size, seq_len, hidden_dim)
        return self.fc(output) # Returns (batch_size, seq_len, hidden_dim) - synthetic embedded sequence

class Discriminator(nn.Module):
    def __init__(self, hidden_dim):
        super(Discriminator, self).__init__()
        # Discriminator GRU processes sequences in the hidden_dim space
        self.gru = nn.GRU(hidden_dim, hidden_dim, batch_first=True)
        # Final FC layer for binary classification based on the last hidden state
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, h): # h: (batch_size, seq_len, hidden_dim)
        output, _ = self.gru(h) # output: (batch_size, seq_len, hidden_dim)
        # Take the last output of the GRU for classification
        return self.sigmoid(self.fc(output[:, -1, :])) # (batch_size, 1)


# Initialize models
embedder = Embedder(feature_dim, hidden_dim)
recovery = Recovery(hidden_dim, feature_dim) # Corrected init: removed seq_len
generator = Generator(noise_dim, hidden_dim) # Corrected init: removed feature_dim
discriminator = Discriminator(hidden_dim)

# Initialize optimizers (adjust learning rates as per TimeGAN paper)
optimizer_E = optim.Adam(embedder.parameters(), lr=gan_args['lr'])
optimizer_G = optim.Adam(generator.parameters(), lr=gan_args['lr'])
optimizer_D = optim.Adam(discriminator.parameters(), lr=gan_args['lr'])

# Define loss functions (e.g., Binary Cross Entropy for GAN, MSE for reconstruction)
bce_loss = nn.BCELoss()
mse_loss = nn.MSELoss()

print("PyTorch TimeGAN model components initialized.")
```

This code snippet sets up the basic neural network modules and optimizers. The actual training loop for TimeGAN is complex, involving multiple stages and specific loss functions (reconstruction, unsupervised, supervised). You would typically implement this training loop based on the TimeGAN paper's methodology or an existing PyTorch implementation.

train timegan

```
from torch.utils.data import DataLoader, TensorDataset

# Assuming data_tensor is already defined and contains your preprocessed time series data
# data_tensor shape: (num_samples, seq_len, feature_dim)

# Create a TensorDataset and DataLoader
dataset = TensorDataset(data_tensor)
dataloader = DataLoader(dataset, batch_size=gan_args['batch_size'], shuffle=True)

# Move models to the appropriate device (CPU or GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
embedder.to(device)
recovery.to(device)
generator.to(device)
discriminator.to(device)

# --- Training Loop ---
num_epochs = 100 # Example number of epochs

print(f"Starting TimeGAN training on {device}...")

for epoch in range(num_epochs):
    # Phase 1: Train Autoencoder (Embedding and Recovery)
    # Objective: Minimize reconstruction error between original and recovered data
    # This phase aims to learn an effective embedding and recovery mechanism
    embedder.train()
    recovery.train()
    optimizer_E.zero_grad()

    # Example: forward pass for autoencoder (simplified)
    real_data_batch = next(iter(dataloader))[0].to(device) # Get a batch
    embedded_data = embedder(real_data_batch)
    recovered_data = recovery(embedded_data) # Note: this 'recovery' is simplified in the class definition

    # Reconstruction Loss
    e_loss = mse_loss(recovered_data, real_data_batch)
    e_loss.backward()
    optimizer_E.step()

    # Phase 2: Train Generator (G)
    # Objective: Make the generated sequences indistinguishable from real sequences (in the embedded space)
    # and preserve temporal dynamics.
    generator.train()
    optimizer_G.zero_grad()

    # Generate noise
    noise = torch.randn(gan_args['batch_size'], seq_len, noise_dim).to(device)

    # Generate synthetic data in the embedded space
    synthetic_embedded_data = generator(noise)

    # Embed real data for discriminator input
    embedded_real = embedder(real_data_batch.detach()) # Detach to prevent gradients flowing to E

    # Discriminator's output on embedded synthetic data
    g_loss_unsupervised = bce_loss(discriminator(synthetic_embedded_data), torch.ones(gan_args['batch_size'], 1).to(device))

    # Additional Generator losses (e.g., supervised, moment matching) would go here.
    # For simplicity, using only unsupervised loss here.
    g_loss = g_loss_unsupervised
    g_loss.backward()
    optimizer_G.step()

    # Phase 3: Train Discriminator (D)
    # Objective: Distinguish between embedded real data and embedded synthetic data.
    discriminator.train()
    optimizer_D.zero_grad()

    # Generate new synthetic data in embedded space for discriminator (important for stability)
    noise_d = torch.randn(gan_args['batch_size'], seq_len, noise_dim).to(device)
    synthetic_embedded_data_d = generator(noise_d).detach() # Detach synthetic data

    # Embed real data for discriminator
    embedded_real_d = embedder(real_data_batch.detach())

    # Discriminator loss for real and synthetic data
    d_loss_real = bce_loss(discriminator(embedded_real_d), torch.ones(gan_args['batch_size'], 1).to(device))
    d_loss_fake = bce_loss(discriminator(synthetic_embedded_data_d), torch.zeros(gan_args['batch_size'], 1).to(device))

    d_loss = d_loss_real + d_loss_fake
    d_loss.backward()
    optimizer_D.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch {epoch+1}/{num_epochs}, E Loss: {e_loss.item():.4f}, G Loss: {g_loss.item():.4f}, D Loss: {d_loss.item():.4f}")

print("TimeGAN training complete.")
```

convert back to original datashape

```
import torch

# Assuming device, generator, recovery, seq_len, noise_dim, and data_tensor are defined from previous cells

# Ensure models are in evaluation mode
generator.eval()
recovery.eval()

# Generate noise for synthetic data
# We generate as many synthetic samples as the original dataset had (num_samples * seq_len)
num_synthetic_sequences = data_tensor.shape[0]
synthetic_noise = torch.randn(num_synthetic_sequences, seq_len, noise_dim).to(device)

# Generate synthetic embedded data using the trained generator
with torch.no_grad():
    synthetic_embedded_data_eval = generator(synthetic_noise)

# Recover synthetic data to original feature space using the trained recovery
with torch.no_grad():
    # The recovery module outputs (batch_size, seq_len, feature_dim)
    synthetic_data_recovered_tensor = recovery(synthetic_embedded_data_eval)
    # Convert to numpy and move to CPU
    synthetic_data = synthetic_data_recovered_tensor.cpu().numpy()

# Now, synthetic_data is a (num_synthetic_sequences, seq_len, feature_dim) numpy array
# Reshape it to (total_timesteps, feature_dim) as expected by inverse_transform
synthetic = synthetic_data.reshape(-1, feature_dim) # feature_dim is 5

synthetic = scaler.inverse_transform(synthetic)

synthetic_df = pd.DataFrame(
    synthetic,
    columns=df.columns
)

print(synthetic_df.head())
```
